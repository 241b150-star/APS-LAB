# MATRIX CHAIN MULTIPLICATION Dynamic Programming with table filling.
Problem Description:
Find minimum number of scalar multiplications needed to multiply a chain of matrices..

#Theory
Cost of multiplying two matrices:
If A is p×q and B is q×r, cost = p × q × r

Dynamic Programming idea:
Break problem into smaller subproblems.
Store results in a table to avoid recomputation.

m[i][j] = minimum cost to multiply Ai to Aj

## Algorithm
1. Take array p[] of size n
2. Create table m[n][n]
3. Set m[i][i] = 0
4. Fill table diagonally
5. For each chain length L:
    Compute m[i][j]
    Try all possible k between i and j
6. Choose minimum cost
7. Return m[1][n-1]

```python
import sys
def matrix_chain_multiplication(p):
    n = len(p)

    # Create DP table for costs
    m = [[0 for _ in range(n)] for _ in range(n)]
    # Create DP table for split points
    s = [[0 for _ in range(n)] for _ in range(n)]

    # L is chain length
    for L in range(2, n):
        for i in range(1, n - L + 1):
            j = i + L - 1
            m[i][j] = sys.maxsize

            for k in range(i, j):
                cost = (m[i][k] +
                        m[k+1][j] +
                        p[i-1] * p[k] * p[j])

                if cost < m[i][j]:
                    m[i][j] = cost
                    s[i][j] = k  # Store the optimal split point

    return m, s
```

```python
p = [5, 10, 3, 12, 5, 50, 6]
m_table, s_table = matrix_chain_multiplication(p)
optimal_cost = m_table[1][len(p)-1]
print("Minimum multiplication cost:", optimal_cost)
print("\nDP Table (Minimum Costs):")
for row in m_table[1:]:
    print(row[1:])
def print_optimal_parenthesization(s, i, j):
    if i == j:
        print(f"A{i}", end="")
    else:
        print("(", end="")
        print_optimal_parenthesization(s, i, s[i][j])
        print(")(", end="")
        print_optimal_parenthesization(s, s[i][j] + 1, j)
        print(")", end="")
print("\nOptimal Parenthesization:", end=" ")
print_optimal_parenthesization(s_table, 1, len(p) - 1)
print()
def calculate_naive_cost(p_dims):
    num_matrices = len(p_dims) - 1
    if num_matrices <= 1:
        return 0
    naive_cost = 0
    naive_cost += p_dims[0] * p_dims[1] * p_dims[2]
    current_rows = p_dims[0]
    current_cols = p_dims[2]
    for i in range(3, num_matrices + 1):
        naive_cost += current_rows * current_cols * p_dims[i]
        current_cols = p_dims[i]
    return naive_cost
naive_cost = calculate_naive_cost(p)
print(f"\nNaive multiplication cost (left-to-right): {naive_cost}")
multiplications_saved = naive_cost - optimal_cost
print(f"Multiplications saved: {multiplications_saved}")
```
