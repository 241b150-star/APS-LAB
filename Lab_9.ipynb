# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):
    distances = {}
    predecessors = {}
    visited = set()
    for node in graph:
        distances[node] = float('inf')
        predecessors[node] = None
    distances[start] = 0
    while len(visited) < len(graph):
        min_node = None
        min_distance = float('inf')
        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node
        if min_node is None:
            break
        visited.add(min_node)
        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight
            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                predecessors[neighbor] = min_node
    return distances, predecessors
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```

Now, let's implement
1. Modify program for user input graph using dictionary.
2. Display shortest path with total distance.
3. Add two more nodes in graph.
4. Test algorithm with different source nodes.

```python
def get_user_graph_input():
    graph = {}
    num_nodes = int(input("Enter the number of nodes in the graph: "))

    for i in range(num_nodes):
        node = input(f"Enter node name {i+1}: ")
        graph[node] = {}

        num_neighbors = int(input(f"How many neighbors does node {node} have? "))
        for j in range(num_neighbors):
            neighbor = input(f"Enter neighbor {j+1} of {node}: ")
            weight = int(input(f"Enter weight of edge from {node} to {neighbor}: "))
            graph[node][neighbor] = weight
    return graph
```

```python
def print_shortest_path(distances, predecessors, start_node, target_node):
    if target_node not in distances or distances[target_node] == float('inf'):
        print(f"No path found from {start_node} to {target_node}.")
        return

    path = []
    current = target_node
    while current is not None:
        path.insert(0, current)
        current = predecessors[current]

    print(f"Shortest path from {start_node} to {target_node}: {' -> '.join(path)}")
    print(f"Total distance: {distances[target_node]}")
```

```python
new_graph = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 5, 'D': 10, 'E': 3},
    'C': {'A': 2, 'B': 5, 'D': 3, 'F': 8},
    'D': {'B': 10, 'C': 3, 'E': 7, 'F': 1},
    'E': {'B': 3, 'D': 7},
    'F': {'C': 8, 'D': 1}
}
print("--- Testing with new_graph ---")
print("\nSource Node: A")
dist_A, pred_A = dijkstra(new_graph, 'A')
print("Distances from A:", dist_A)
print_shortest_path(dist_A, pred_A, 'A', 'E')
print_shortest_path(dist_A, pred_A, 'A', 'F')
print("\nSource Node: B")
dist_B, pred_B = dijkstra(new_graph, 'B')
print("Distances from B:", dist_B)
print_shortest_path(dist_B, pred_B, 'B', 'A')
print_shortest_path(dist_B, pred_B, 'B', 'F')

# Test with source node 'C'
print("\nSource Node: C")
dist_C, pred_C = dijkstra(new_graph, 'C')
print("Distances from C:", dist_C)
print_shortest_path(dist_C, pred_C, 'C', 'E')
print_shortest_path(dist_C, pred_C, 'C', 'A')

# Test with source node 'F'
print("\nSource Node: F")
dist_F, pred_F = dijkstra(new_graph, 'F')
print("Distances from F:", dist_F)
print_shortest_path(dist_F, pred_F, 'F', 'A')
print_shortest_path(dist_F, pred_F, 'F', 'B')
```
